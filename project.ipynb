# DROP THE UNNAMED COLUMN

```python
df=df.drop("Unnamed: 0",axis=1)
print(df.head)
```

# CORRECTING WEEKLY STYDY HOURS(transformation)

```python
df["WklyStudyHours"]=df["WklyStudyHours"].str.replace("10-May","5-10")
df
```

# Gender Distribution using seaborn (countplot)

```python
plt.figure(figsize=(4,4)) # for the measurements of chart
ax=sb.countplot(data = df,x="Gender") # to count the male and female in the dataframe using countplot
ax.bar_label(ax.containers[0]) #to display the exact count here we used BAR_LABEL AND CONTAINERS
plt.show() #to display the chart
```

# Getting analysis of student's avg based on the no of hours they study

```python
gb=df.groupby("WklyStudyHours").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"}) # grouping based on the weekly study hours 
print(gb)
plt.figure(figsize=(5,4)) # to alter the measurement of the chart
sb.heatmap(gb,annot=True) # to display the chart based on Variable GB here we used heatmap() and annot is used to get the specific values( for clear explanation remove annot and check once)
plt.title("     Students Score based on Weekly Study Hours")
plt.show()
```

from the above chart its clear that the students who study 5 - 10 hoyrs a week scored more than the students who studied less than 5 HOURS 

# Getting analysis of student's avg based on their Parents Educational Background

```python
gb1=df.groupby("ParentEduc").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
print(gb1)
plt.figure(figsize=(5,5))
sb.heatmap(gb1,annot=True) # THE SAME CONCEPT AS THE PREVIOUS SECTION
plt.title("Students performance based on Parents educational background")
plt.show()
```

from the above analysis we can see that the parents educational backgroung has some impact in the performance of their children."HOWEVER ITS CAN COMPLETELY TRUE IN REAL LIFE" 

# Students Performance based on Completion of Test preparation

```python
gb3=df.groupby("TestPrep").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"}) # grouping based on the weekly study hours 
print(gb3)
plt.figure(figsize=(5,4)) # to alter the measurement of the chart
sb.heatmap(gb3,annot=True) # to display the chart based on Variable GB here we used heatmap() and annot is used to get the specific values( for clear explanation remove annot and check once)
plt.title("     Students Score based on Test Preparations")
plt.show()
```

Here we can clearly notice that the students who did not complete test preparation are scored lesser than the students who completed their test preparation


# Comparison of female and male students score in math,write,read

```python
plt.figure(figsize=(5,3))
gb4=df.groupby("Gender").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
sb.heatmap(data=gb4,annot=True)
plt.title("Comparison based on gender")
plt.show()
```

From the above analysis we can clearly see that female students performance is better than male students

# Analysis on Ethnic Groups of Students using PANDAS (PERCENTAGE AND PIE CHART)

```python
groupA=df.loc[(df["EthnicGroup"]=="group A")].count()
groupB=df.loc[(df["EthnicGroup"]=="group B")].count()
groupC=df.loc[(df["EthnicGroup"]=="group C")].count()
groupD=df.loc[(df["EthnicGroup"]=="group D")].count()
groupE=df.loc[(df["EthnicGroup"]=="group E")].count()
l=["group A","group B","group C","group D","group E"]
mylist=[groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]
print(mylist)
plt.pie(mylist,labels=l,autopct="%1.2f%%")
plt.show()
```

Here we used pandas(df.loc) to get data of ethnic gropu's each categories,then plt.pie() for the piechart,then AUTOPCT=%1.2F%% to display the count in percentage and also labels to display the names
